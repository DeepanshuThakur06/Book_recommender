## 3. ***Data Wrangling***

### Data Wrangling Code

#### Popularity based Recommendar system


```python
# Write your code to make your dataset analysis ready.
ratings_with_name=ratings.merge(books, on='ISBN')
```

```python
num_ratings_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns={'Book-Rating':'Num-rating'},inplace=True)
num_ratings_df
```

```python
# Group the ratings_with_name DataFrame by 'Book-Title' and calculate the mean rating for each book
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the column containing the mean rating to 'Avg-Rating'
avg_rating_df.rename(columns={'Book-Rating': 'Avg-Rating'}, inplace=True)

# Display the resulting DataFrame
popular_books_df=num_ratings_df.merge(avg_rating_df,on='Book-Title')
popular_books_df=popular_books_df[popular_books_df['Num-rating']>250]
```

```python
popular_books_df=popular_books_df.sort_values(by='Avg-Rating',ascending=False).head(50)
popular_books_df
```

```python
popular_df=popular_books_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','Publisher','Num-rating']]
```

```python
popular_df
```

```python
import pickle
pickle.dump(popular_df,open('popular_book.pkl','wb'))
```

#### Colabrative based Filtering

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
educated_users=x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

Check the similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
def recommend_system(book_name):
  index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x : x[1],reverse=True)[1:11]
  data=[]
  for i in similar_items:
      item=[]
      temp_df=books[books['Book-Title']==pt.index[i[0]]]
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
      data.append(item)
  return data
```

```python
pickle.dump(pt,open('pt.pkl','wb'))
pickle.dump(books,open('books.pkl','wb'))
pickle.dump(similarity_score,open('similarity_score.pkl','wb'))
```

```python
recommend_system('Me Talk Pretty One Day')
```
